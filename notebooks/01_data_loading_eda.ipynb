```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df, x='Churn')

plt.title('Customer Churn Distribution')
plt.show()
```

### Finding

Most customers have remained with the company, but a significant portion have churned.

This suggests that customer retention is an important business challenge and understanding the factors behind churn could provide meaningful business value.

## Contract Type vs Churn

Contract type can have a significant impact on customer retention. In this section, I want to explore whether customers on different contract plans have different churn rates.

```python
df['Contract'].value_counts()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x='Contract', hue='Churn')

plt.title('Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Customer Count')

plt.show()
```

```python
contract_churn = pd.crosstab(
    df['Contract'],
    df['Churn'],
    normalize='index'
) * 100

contract_churn.round(2)
```

```python
contract_churn.round(2)
```

## Tenure vs Churn

Customer tenure represents how long a customer has been with the company.

In this section, I want to explore whether customer retention improves as tenure increases.

```python
plt.figure(figsize=(10,5))

sns.histplot(
    data=df,
    x='tenure',
    hue='Churn',
    multiple='stack'
)

plt.title('Tenure Distribution by Churn')
plt.xlabel('Tenure (Months)')
plt.ylabel('Customer Count')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Churn',
    y='tenure'
)

plt.title('Tenure by Churn Status')

plt.show()
```

```python
df.groupby('Churn')['tenure'].mean().round(2)
```

## Monthly Charges vs Churn

Pricing can influence customer retention. In this section, I explore whether customers with higher monthly charges are more likely to leave the service.


```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Churn',
    y='MonthlyCharges'
)

plt.title('Monthly Charges by Churn Status')

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    data=df,
    x='MonthlyCharges',
    hue='Churn',
    multiple='stack'
)

plt.title('Monthly Charges Distribution by Churn')
plt.xlabel('Monthly Charges')
plt.ylabel('Customer Count')

plt.show()
```

```python
df.groupby('Churn')['MonthlyCharges'].mean().round(2)
```
